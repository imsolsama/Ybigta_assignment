```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/YBIGTA/24-Winter Session/0116_통계기초_회귀분석/diabetes_XY.csv', index_col = 0)
data.head()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python


#1 - formula 명시하지 않는 방법
Y = data.iloc[:, -1]
X = data.iloc[:, :-1]
X = sm.add_constant(X)

model = sm.OLS(Y, X)
res = model.fit()

print(res.summary())
```

```python
#2 - R처럼 formula 사용하기 (다중회귀에서 변수 변경할 때 보다 편리함)
form_mod = ols('disease_progression ~ age + sex + bmi + bp + s1 + s2 + s3 + s4 + s5 + s6', data = data)
res2 = form_mod.fit()

print(res2.summary())
```

```python
# 산점도 행렬 그리기 - 다중공선성 확인
X_ = X.iloc[:, 1:]
pd.plotting.scatter_matrix(X_)
plt.show()
```

```python
# 일반 산점도 한꺼번에 그리기

def draw_scatter(ii, jj, title):
    X_cols = np.array(X_.columns).reshape(ii,-1)
    fig, axs = plt.subplots(ii,jj, figsize = (10,10), layout = 'constrained')
    fig.suptitle(title)
    axs_list = [(i,j) for i in range(0,ii) for j in range(0,jj)]
    for ax in axs_list:
        axs[ax].set_title(X_cols[ax])
        axs[ax].scatter(X_[X_cols[ax]], Y)
    plt.show()

draw_scatter(5, 2, "Scatter Plot Between Explanatory and Response Variables")
```

```python
#잔차도 + 다른 여러 그림 그리기
fig = plt.figure(figsize=(10, 8))
fig = sm.graphics.plot_regress_exog(form_mod.fit(), 's1', fig=fig)
```

```python
#잔차 히스토그램
residuals = Y - res.predict()
plt.hist(residuals, bins = 20)
plt.show()
```

```python
#다중공선성 확인
vif = pd.DataFrame()
vif['VIF Factors'] = [variance_inflation_factor(X_, i)for i in range(X_.shape[1])]
vif['features'] = X_.columns
vif['is_vif'] = vif['VIF Factors'] > 10
vif = vif.sort_values("VIF Factors").reset_index(drop=True)
vif
```

```python
X = data['Study_Hours']
y = data['PI']

X = sm.add_constant(X)

model = sm.OLS(y, X)
results = model.fit()

print(results.summary())
```

2.3. (2.1)과 (2.2)의 분석을 비교하고 Study_Hours와 PI의 관계를 짧게, 자유롭게 설명해보세요.

2.1의 다중선형회귀분석과 2.2의 단순선형회귀분석식의 OLS SUMMARY를 봐보면 기본적으로 두 모델다 P-value가 0.00이기에 통계적으로는 유의미하다는것을 알 수 있지만 R-Squared값은 2.1의 모델이 훨씬 높은것으로 보아하니 'Study_Hours' 뿐만 아니라 다른 변수들도 함께 고려할 때 'PI'의 변동성을 더 잘 설명하고 있다는 것을 분석할 수 있다, 'Study_Hours'와 'PI'의 관계를 설명하자면, 'Study_Hours'가 증가할수록 'PI'도 증가하는 추세이다 또한, 'Study_Hours'의 회귀계수가 양수이므로, 학습시간이 늘어날수록 성적이 개선되는 것으로 해석가능. 하지만, 2.2의 단순선형회귀분석에서 보여준 것처럼, 'Study_Hours'만으로 'PI'를 완벽하게 예측하기는 어렵다고 생각한다

산점도를 확인해보면 "Sutdy_Hours"에 따라 PI가 오르는것을 확인할 있음

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from statsmodels.graphics.gofplots import ProbPlot

sns.scatterplot(x='Study_Hours', y='PI', data=data)
plt.title('Scatter plot of Study_Hours vs PI')
plt.show()

model = smf.ols('PI ~ Study_Hours', data=data).fit()

sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})
plt.title('Residual plot')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.show()

QQ = ProbPlot(model.resid)
QQ.qqplot(line='45', alpha=0.5, color='#4C72B0', lw=1)
plt.title('Normal Q-Q')
plt.show()
```
